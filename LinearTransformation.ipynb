# Transpose of a matrix:  



```python
import numpy as np
import matplotlib.pyplot as plt

def plot_matrix_and_its_transpose(matrix):
    # Transposing the matrix
    transpose_matrix = matrix.T

    # Creating a figure and axis (this line is used to create a figure with a grid of subplots)
    fig, ax = plt.subplots(1, 2, figsize=(10, 5))

    # Plotting the original matrix
    cax1 = ax[0].matshow(matrix, cmap='Blues')
    ax[0].set_title('Original Matrix')

    # Plotting the transpose of the matrix
    cax2 = ax[1].matshow(transpose_matrix, cmap='Blues')
    ax[1].set_title('Transposed Matrix')

    plt.show()

# Example matrix
matrix = np.array([[1, 2, 3], [4, 5, 6]])

# Plotting the matrix and its transpose
plot_matrix_and_its_transpose(matrix)
```

# Linear Transformation

Let's understand first what do we mean by a transformation. It is just another word for a function. Transformations are functions that take in some vector (matrix) and return us a vector(matrix)

Then why do we call it a transformation?
For visualizing what they do.

Visualization of what a linear transformation is :

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_linear_transformation(matrix, points):
    """
    Apply a linear transformation to a set of points and plot the original and transformed shapes.

    :param matrix: A 2x2 transformation matrix.
    :param points: A set of points (2D) to be transformed.
    """
    # Applying the linear transformation
    transformed_points = np.dot(matrix, points)

    # Plotting the original shape
    plt.figure()
    plt.fill(points[0], points[1], 'blue', alpha=0.3, label='Original Shape')

    # Plotting the transformed shape
    plt.fill(transformed_points[0], transformed_points[1], 'red', alpha=0.3, label='Transformed Shape')

    plt.legend()
    plt.grid(True)
    plt.axhline(0, color='black',linewidth=0.5)
    plt.axvline(0, color='black',linewidth=0.5)
    plt.title("Linear Transformation")
    plt.show()

# Example points representing a square
square_points = np.array([[0, 1, 1, 0, 0],
                          [0, 0, 1, 1, 0]])

# Example transformation matrix (e.g., rotation matrix)
angle = np.pi / 4  # 45 degrees
transformation_matrix = np.array([[np.cos(angle), -np.sin(angle)],
                                  [np.sin(angle),  np.cos(angle)]])

# Plotting the transformation
plot_linear_transformation(transformation_matrix, square_points)
```


Now let's talk about the properties of linear transformations:
What are the properties that make a transformation a linear transformation?
The two properties : (Additivity and Homogeneity)

## Additivity:
The additivity of a linear transformation of a matrix means that the transformation preserves addition, i.e., T(A + B) = T(A) + T(B) for any matrices A and B, where T is the linear transformation. Visualizing this:

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_additivity(matrix, points_a, points_b):
    # Transforming individual sets of points
    transformed_a = matrix @ points_a ## you can also use np.dot(matrix, points)
    transformed_b = matrix @ points_b ## you can also use np.dot(matrix, points)

    # Transforming the sum of sets of points
    sum_of_points = points_a + points_b
    transformed_sum = matrix @ sum_of_points ## you can also use np.dot(matrix, points)

    # Plotting the original shapes and their transformations
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Plot for T(A + B)
    axs[0].fill(transformed_sum[0], transformed_sum[1], 'orange', alpha=0.3, label='T(A + B)')
    axs[0].set_title('T(A + B)')
    axs[0].grid(True)
    axs[0].axhline(0, color='black', linewidth=0.5)
    axs[0].axvline(0, color='black', linewidth=0.5)

    # Plot for T(A) + T(B)
    axs[1].fill(transformed_a[0] + transformed_b[0], transformed_a[1] + transformed_b[1], 'purple', alpha=0.3, label='T(A) + T(B)')
    axs[1].set_title('T(A) + T(B)')
    axs[1].grid(True)
    axs[1].axhline(0, color='black', linewidth=0.5)
    axs[1].axvline(0, color='black', linewidth=0.5)

    plt.legend()
    plt.suptitle("Additivity of Linear Transformation")
    plt.show()

# Example points for shapes A and B
points_a = np.array([[0, 2, 2, 0, 0], [0, 0, 2, 2, 0]])  # Square
points_b = np.array([[2, 4, 4, 2, 2], [0, 0, 2, 2, 0]])  # Adjacent Square

# Example transformation matrix
transformation_matrix = np.array([[1, -0.5], [0.5, 1]])

# Plotting additivity with subplots
plot_additivity(transformation_matrix, points_a, points_b)
```

## Homogeneity

The homogeneity of a linear transformation of a matrix means that the transformation scales uniformly with respect to scalar multiplication, i.e., T(k * A) = k * T(A) for any matrix A and scalar k, where T is the linear transformation.





```python
import numpy as np
import matplotlib.pyplot as plt

def plot_homogeneity(matrix, points, scalar):
    # Transforming the set of points
    transformed_points = matrix @ points ## you can also use np.dot(matrix, points)

    # Applying scalar multiplication after transformation
    scaled_transformed = scalar * transformed_points

    # Applying scalar multiplication before transformation
    transformed_scaled = matrix @ (scalar * points) ## you can also use np.dot(matrix, points)

    # Plotting the transformations
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Plot for T(k * A)
    axs[0].fill(scaled_transformed[0], scaled_transformed[1], 'red', alpha=0.3, label='T(k * A)')
    axs[0].set_title('T(k * A)')
    axs[0].grid(True)
    axs[0].axhline(0, color='black', linewidth=0.5)
    axs[0].axvline(0, color='black', linewidth=0.5)

    # Plot for k * T(A)
    axs[1].fill(scalar * transformed_points[0], scalar * transformed_points[1], 'green', alpha=0.3, label='k * T(A)')
    axs[1].set_title('k * T(A)')
    axs[1].grid(True)
    axs[1].axhline(0, color='black', linewidth=0.5)
    axs[1].axvline(0, color='black', linewidth=0.5)

    plt.legend()
    plt.suptitle("Homogeneity of Linear Transformation")
    plt.show()

# Scalar value
c = 1.5

# Plotting homogeneity with subplots
plot_homogeneity(transformation_matrix, points_a, c)
```
