Graphing 2-dimensional equations using Python

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the 2D equation
def equation(x):
    return x**2 + 2*x + 1

# Generate x values
# numpy.linspace is a function from the NumPy library in Python,
# and it is used to create an array of evenly spaced values over a specified range.
x_values = np.linspace(-10, 10, 100)

# Calculate y values using the equation
y_values = equation(x_values)

# Plot the equation
plt.plot(x_values, y_values, label='y = x^2 + 2x + 1') #update it along with your function
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of the 2D Equation')
plt.legend()
plt.grid(True)
plt.show()
```

Graphing 3-dimensional equations using Python

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the 3D equation as a string
equation_str = 'z = x^2 + y^2'

# Define the 3D equation as a Python function
def equation(x, y):
    return x**2 + y**2

# Generate x and y values using linspace to get evenly spaced values
x_values = np.linspace(-10, 10, 100)
y_values = np.linspace(-10, 10, 100)

#Creating the 3D figure from X and Y values
X, Y = np.meshgrid(x_values, y_values)
Z = equation(X, Y)

#Visualization part of the code
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis') #viridis is the colormap that we use

# Set labels -> just to make the visualization prettier!
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Graph of the 3D Equation')

plt.show()
```

# **(Optional) If you want to understand how meshgrid works.**

The meshgrid function in NumPy helps with the creation of coordinate grids for plotting. Given input arrays representing coordinates along different axes, meshgrid generates 2D or 3D arrays containing all possible combinations of these coordinates.
Consider 1-D arrays x = [1, 2, 3] and y = [4, 5],
now using meshgrid(x, y) we can produce two arrays, X and Y, where each element is a combination of an x and y coordinate.
