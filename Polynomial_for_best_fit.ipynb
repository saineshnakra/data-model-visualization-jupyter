#### Line of Best Fit (Linear Regression)

A line of best fit is a straight line that best represents the data on a scatter plot. This line may pass through some of our desired points, none of the points, or all of the points.


**Step 1**: Gather or define a dataset.

**Step 2**: Calculate the mean of x-values and y-values.

**Step 3**: Calculate the slope (m) and intercept (b) of the line of best fit.

(In linear regression (discussed in previous notebooks), we often compute this line using the least squares method, minimizing the sum of the squares of the differences between the observed values and the values predicted by the line.
)

**Step 4**: Use the slope and intercept to predict new values



**(Optional: understanding the least square method)**

How are we calculating the slope (m)


```
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)

```



**(x - x_mean)** computes the difference between each x-value and the mean of all x-values. This represents the deviation of each x-value from the mean.

**(y - y_mean)** computes the difference between each y-value and the mean of all y-values. This represents the deviation of each y-value from the mean.

**(x - x_mean) * (y - y_mean)** calculates the product of the deviations.

**np.sum((x - x_mean) * (y - y_mean))** sums these products across all data points.

**((x - x_mean)^2)** computes the square of the deviation of each x-value from its mean, which is the formula for variance of x.

**np.sum((x - x_mean)^2)** sums the squared deviations, which gives the total variance of x for the given dataset.

The division of the covariance by the variance gives the slope (m) of the regression line, indicating how much y is expected to increase when x increases by one unit.

````python
import numpy as np

# Step 1: Define a dataset
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 3, 5, 7, 11])

# Step 2: Mean of x and y
x_mean = np.mean(x)
y_mean = np.mean(y)

# Step 3: Calculate slope (m) and intercept (b) [using least squares method]
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)
b = y_mean - m * x_mean

# Step 4: Predict new values
y_pred = m * x + b

print("Slope:", m)
print("Intercept:", b)
print("Predicted y values:", y_pred)
````
